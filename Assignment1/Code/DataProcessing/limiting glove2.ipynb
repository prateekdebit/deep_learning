```python
import os,sys
from scipy import stats
import numpy as np
f1 = open('Holmes.machine_format.questions.number_format_w.txt', 'r+')
f2 = open('glove.840B.300d.txt', 'r')
f3 = open('new_glove.txt', 'w')



print('sucess')


```

```python
mapping = []
for i in range(0,2195990):
    mapping.append(0)
glove = f2.readlines()
k = 0
for l1 in f1:
    k = k+1
    line1 = l1[:-1]
    strings1 = line1.split()
    length = len(strings1)
    for i in range(0,length):
        ids = int(strings1[i])
        #print(ids)
        if mapping[ids]==0 :
            
            f3.write(glove[ids])
            #f3.write('\n')
            mapping[ids] = 1
    if k%50000==0:
        print(k)

print('finally sucess')
```
